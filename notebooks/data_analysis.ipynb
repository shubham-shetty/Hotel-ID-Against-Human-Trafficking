Number of unique hotel IDs - Number of classification labels

```python
df.shape
hotel_id_counts_df = (df['hotel_id'].value_counts()).reset_index()
hotel_id_counts_df.head()
print(len(hotel_id_counts_df.index))
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(5,2))
hotel_id_counts_df["hotel_id"].plot()
fig.suptitle('Number of images per hotel ID')
plt.xlabel('Hotel IDs')
plt.ylabel('Count')
```

```python
chain_id_counts_df = df['chain'].value_counts().sort_index().to_frame()
chain_id_counts_df.head()
fig = plt.figure(figsize=(25,8))
chain_id_counts_df["chain"].plot()
fig.suptitle('Number of images per chain')
plt.xlabel('Chain IDs')
plt.ylabel('Count')
```

```python
hotels_per_chain_df = df.groupby(["chain"]).agg({"hotel_id": [pd.Series.nunique]}).sort_index()
hotels_per_chain_df.columns = ["_".join(x) for x in hotels_per_chain_df.columns.ravel()]
fig = plt.figure(figsize=(5,2))
hotels_per_chain_df["hotel_id_nunique"].plot()
fig.suptitle('Number of hotels per chain')
plt.xlabel('Chain IDs')
plt.ylabel('Count')
```
